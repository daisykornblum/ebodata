## Replacement contractions, removal stopwords, lemmatization

```python
%matplotlib inline
from sklearn.feature_extraction.text import CountVectorizer
import nltk
#nltk.download()
from nltk.corpus import stopwords
from nltk import word_tokenize
from nltk.stem import PorterStemmer
from nltk.stem.wordnet import WordNetLemmatizer
from nltk import pos_tag

import numpy as np
import pandas as pd
from pymongo import MongoClient
```

```python
import re

punctuation_regex =  re.compile(ur'[\.,\*\&\^\%\$£!\?\":;\#\+=\-_\'\|]+', re.UNICODE)

replacement_patterns = [
  (r'won\'t', 'will not'),
  (r'can\'t', 'cannot'),
  (r'i\'m', 'i am'),
  (r'ain\'t', 'is not'),
  (r'(\w+)\'ll', '\g<1> will'),
  (r'(\w+)n\'t', '\g<1> not'),
  (r'(\w+)\'ve', '\g<1> have'),
  (r'(\w+)\'s', '\g<1> is'),
  (r'(\w+)\'re', '\g<1> are'),
  (r'(\w+)\'d', '\g<1> would')
]

class RegexpReplacer(object):
  def __init__(self, patterns=replacement_patterns):
    self.patterns = [(re.compile(regex), repl) for (regex, repl) in 
      patterns]

  def replace(self, text):
    s = text
    for (pattern, repl) in self.patterns:
      s = re.sub(pattern, repl, s)
    return s

def bag_of_words(words):
  return dict([(word, True) for word in words])
```

```python
def _connect_mongo(host, port, username, password, db):
    """ A util for making a connection to mongo """

    if username and password:
        mongo_uri = 'mongodb://%s:%s@%s:%s/%s' % (username, password, host, port, db)
        conn = MongoClient(mongo_uri)
    else:
        conn = MongoClient(host, port)


    return conn[db]

def read_mongo(db, collection, query={}, fields=None, limit_to=None, host='localhost', port=27017, username=None, password=None, no_id=True):
    """ Read from Mongo and Store into DataFrame """

    # Connect to MongoDB
    db = _connect_mongo(host=host, port=port, username=username, password=password, db=db)

    # Make a query to the specific DB and Collection
    cursor = db[collection].find(query, fields)
    
    if limit_to is not None:
        cursor = cursor.limit(limit_to)

    # Expand the cursor and construct the DataFrame
    df =  pd.DataFrame(list(cursor))

    # Delete the _id
#     if no_id:
#         del df['_id']

    return df
```

```python
deduped_data = read_mongo('ebodata', 'deduped_data_20141119')
deduped_data.head(5)
print 'num: ' + repr(len(deduped_data))
```

```python
wnl = WordNetLemmatizer()
english_stops = set(stopwords.words('english'))
replacer = RegexpReplacer()
wordnet_tags = ['n', 'v']

def lemmatize(token, tag):
    if tag[0].lower() in ['n', 'v']:
        return wnl.lemmatize(token, tag[0].lower())
    return token

# print 'num: ' + repr(len(data))

lemmed_data = []
data = deduped_data[:10000]
allwords = []
for row in data.iterrows():
    data_row = {}
    data_row['_id'] = row[1][0]
    data_row['id_str'] = row[1][2]
    data_row['cleaned_text'] = row[1][1]
### Regular expression word replacement common contractions
    data_row['text_rep_contrax']= replacer.replace(data_row['cleaned_text'])

### Regular expression removal punctuation from text
    data_row['text_no_punct'] = punctuation_regex.sub('', data_row['text_rep_contrax'])
    
### Stopword removal from text    
    data_row['words'] = [w for w in data_row['text_no_punct'].split()]
    data_row['words_no_stops'] = [w for w in data_row['words'] if not w in stopwords.words('english')]
    
### Part of Speech tags    
    
    data_row['tagged'] = [pos_tag(word_tokenize(w)) for w in data_row['words_no_stops']]#
#     print data_row['tagged']
    
### Lemmatization remaining words      
    lemmaArray = []
#     print row[1][2]
    wordsArray = data_row['tagged']
#     print 'num: ' + repr(len(wordsArray))
#     print wordsArray
    for word in wordsArray:
        lemma = [lemmatize(token, tag) for token, tag in word]
        lemmaArray.append(lemma[0])
        allwords.append(lemma[0])
#         print lemmaArray        
    data_row['lemmed'] = lemmaArray
    
### Flatten list of words to sentence
    data_row['flat_words'] = " ".join(data_row['lemmed'])
#   print data_row['flat_words']

    lemmed_data.append(data_row)
    
# print 'num: ' + repr(len(data))

    
### Generate dataframe for final variables    
# strippeddata = pd.DataFrame(data)
briefdata = pd.DataFrame(lemmed_data, columns=['_id', 'id_str', 'cleaned_text', 'flat_words'])
briefdata

### Set up input for CountVectorizer
# tryVect = briefdata['flat_words']
# tryVect

# lemmed_data = data

# example = briefdata[:10]
# df = example['flat_words']
# df
# df.to_csv('egflat.csv', sep='\t', encoding='utf-8')
```
